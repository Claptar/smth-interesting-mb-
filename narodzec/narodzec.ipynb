```python
import os
import math
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from laba import*
```

```python
file = 'EL15.xlsx'
data15 = pd.read_excel(file)
data15.fillna(0, inplace=True)
```

```python
data15['вид территориальной единицы'] = data15['Наименование административно- территориальной (территориальной) единицы'].str.contains('г. ')
data15['вид территориальной единицы'].replace(True, 'город', inplace=True)
data15['вид территориальной единицы'].replace(False, 'район', inplace=True)
```

## Распределение явка/процент за Лукашенко А.Г. в сравнении город/район

**Размер метки** соответвсвует количеству (полному) избирателей на участке. Соответсвенно, на графике очень отчётливо видна зависимость: чем больше населённый пункт -- тем меньше явка и процент голосов "за". Так же заметно, что, даже при одинаковом количестве избирателей на районном и городских участках, и явка и процент обычно больше на районных участках.


Хочется отметить "забавную" кучность точек отвечающих районам крупных городов. "Забавнее" всего получилось с районами Витебска: два района почти полностью совпадают, а третий сильно выбивается.

```python
import plotly.express as px
fig = px.scatter(data15,
                 x="Явка (проц)",
                 y="Лукашенко А.Г. (процент)",
                 color='вид территориальной единицы',
                 hover_name='Наименование административно- территориальной (территориальной) единицы',
                title='Выборы 2015-го года',
                width=1000,
                height=600,
                size='Количество избирателей (всего)')
fig.show()
```

Как видно из таблицы ниже кандидат **Лукашенко получил в 1.5 раза больше** голосов избирателей с районов, чем с городов. Так же количество избирателей, **проголосовавших досрочно**, в районах в 1.5 больше аналогичных в городах.

```python
col = ['Количество избирателей (всего)',
       'Количество избирателей, принявших участие в голосовании (всего)',
       'Количество избирателей, принявших участие в голосовании (досрочно)',
       'Количество голосов, поданных против всех кандидатов',
       'Лукашенко А.Г. (кол-во)']
town = data15.groupby(['вид территориальной единицы']).sum()[col]
town['Процент проголосовавших за Лукашенко'] = town['Лукашенко А.Г. (кол-во)'] / town['Количество избирателей (всего)'] * 100
town.head()
```

## Распределение явка/процент за Лукашенко А.Г. в сравнении по областям

В целом единственное, что можно сказать: у Минской области наименьшие медианные значения как явки так и процента "за", Могилёвская область больше всех "рисует" явку, а Гомельская процент.

```python
fig = px.scatter(data15,
                 x="Явка (проц)",
                 y="Лукашенко А.Г. (процент)",
                 color='Область',
                 hover_name='Наименование административно- территориальной (территориальной) единицы',
                 marginal_y="box",
                 marginal_x="box",
                 width=1200,
                 height=800,
                 size='Количество избирателей (всего)'
                )
fig.show()
```
